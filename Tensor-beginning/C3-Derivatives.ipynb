```python
import torch
```

```python
x= torch.tensor(2.0, requires_grad=True)

y = 9*x**4 + 2*x**3 + 3*x**2 + 6*x + 1

y.backward() 
x.grad
```

```python
x = torch.tensor(1.0, requires_grad=True)
z = torch.tensor(2.0, requires_grad=True)
y = x**2 + z**3
y.backward() #compute the gradient, it will take the partical derivative of the function
x.grad #dy/dx
```

```python
z.grad #dy/dz
```
